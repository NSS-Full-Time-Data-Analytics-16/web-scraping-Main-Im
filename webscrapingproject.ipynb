```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
#1a.
senior_python_developer=soup.find('h2', class_='title is-5')
senior_python_developer.text
```

```python
#1b.
all_job_titles=soup.findAll('h2',class_="title is-5")
all_job_titles_list=[]
for i in all_job_titles:
    all_job_titles_list.append(i.text.strip())

all_job_titles_list
```

```python
#1c.
companies=soup.findAll('h3',class_="subtitle is-6 company")
all_job_companies_list=[]
for i in companies:
    all_job_companies_list.append(i.text.strip())

all_job_companies_list
```

```python
locations=soup.findAll('p',class_="location")
all_job_locations_list=[]
for i in locations:
    all_job_locations_list.append(i.text.strip())

all_job_locations_list
```

```python
posting_dates=soup.findAll('p',class_="is-small has-text-grey")
all_job_postings_list=[]
for i in posting_dates:
    all_job_postings_list.append(i.text.strip())

all_job_postings_list
```

```python
#1d
number_one_dataframe=pd.DataFrame({'all_job_titles_list':all_job_titles_list,'all_job_companies_list':all_job_companies_list,'all_job_locations_list':all_job_locations_list,'all_job_postings_list':all_job_postings_list})
```

```python
#2a.
apply_link=soup.findAll('a',string="Apply")
apply_list=[]
for i in apply_link:
    apply_list.append(i['href'].strip())

apply_list
```

```python
number_one_dataframe['link_one']=apply_list
number_one_dataframe
```

```python
#2b
number_one_dataframe['link_part_b']='https://realpython.github.io/fake-jobs/jobs/'+number_one_dataframe['all_job_titles_list'].str.lower().replace(r'\W+','-',regex=True)+'-'+number_one_dataframe.index.astype(str)+'.html'
number_one_dataframe['link_part_b']=number_one_dataframe['link_part_b'].replace(r'-+', '-', regex=True)
number_one_dataframe['link_part_b']
```

```python
#3a
url2 = "https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html"
response2 = requests.get(url2)
soup2 = BS(response2.content, 'html.parser')
description = soup2.find('div', class_='content').find('p').text.strip()
print(description)
```

```python
role_description=[]
for i in number_one_dataframe['link_part_b']:
    url3= i
    response3 = requests.get(url3)
    soup3 = BS(response3.content, 'html.parser')
    description2 = soup3.find('div', class_='content').find('p').text.strip()
    role_description.append(description2.strip())
role_description
```

```python
number_one_dataframe['job_description']=role_description
number_one_dataframe
```

```python
def role_description2(url3):
    response3 = requests.get(url3)
    soup3 = BS(response3.text)
    description2 = soup3.find('div', class_='content').find('p').text.strip()
    return(description2)
```

```python
number_one_dataframe['job_description2'] = number_one_dataframe['link_one'].apply(role_description2)
```
